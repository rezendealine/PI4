**Exercício 7 - Regressão Linear**

A partir das variáveis, calcular o preço estimado do diamante.

Usar:

- Dataset treinamento/testes
- Limpeza dos dados
- Seleção de dados de entrada que maximizam o resultado
- Calcular R2 do modelo

```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/diamonds.csv")
df
```

```python
df_X = df[["carat", "depth", "table"]]
df_Y = df [["price"]]

from sklearn.model_selection import train_test_split

X_treinamento, X_testes, Y_treinamento, Y_testes = train_test_split(df_X, df_Y, test_size=0.2)
```

```python
from sklearn import linear_model

reg = linear_model.LinearRegression()
reg.fit(X_treinamento, Y_treinamento)
reg.coef_
```

```python
Y_pred = reg.predict(X_testes)

from sklearn.metrics import r2_score

r2_score(Y_testes, Y_pred)
```
